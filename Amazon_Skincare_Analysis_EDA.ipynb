```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#displaying dataset from github
product_df1=pd.read_csv("https://github.com/Vakiti/SkinTelligence-Navigating-The-Amazon-Skincare-Universe/raw/datasets/final_dataset.csv")
product_df1
```

```python
'''

In this EDA we will be using single, double and multivariable EDA to understand questions like:

Single Variable EDA:
What is the distribution of product ratings? 
What is the average price of products in your dataset? 
How many products are Prime eligible vs. non-Prime eligible?
What are the top categories of products in your dataset? 
Are there any missing values in the columns? 

Double Variable EDA:

Is there a correlation between product ratings and prices? Plot a scatter plot to visualize the relationship between ratings and prices.

Do Prime eligible products tend to have higher ratings? Compare the average ratings of Prime and non-Prime eligible products.

Are there any patterns between product categories and prices? Explore if certain categories have higher or lower average prices.

How does the number of total reviews relate to ratings? Create a scatter plot to examine if products with more reviews tend to have higher ratings.

Multivariable EDA:

What is the relationship between product ratings, prices, and total reviews across different categories? You can create a heatmap or multiple plots to explore interactions between these variables.
Remember that these are just starting points, and EDA can be highly customized based on your specific goals and questions. The goal of EDA is to uncover patterns, anomalies, and relationships within your dataset that can inform further analysis or decision-making.
'''
```

```python
'''
1) A histogram or box plot can be used to visualize the distribution of ratings.
2) O calculate the mean and median prices and explore the average price distributions of the products.
3) We can create a bar plot or pie chart to show the distribution of prime vs. non-prime products.
4) We can create a frequency table or bar chart to identify the most common categories.
5) This can be done by checking for missing data in each column and exploring how to handle them.
'''

# Step 1: Distribution of Product Ratings (Using a Violin Plot)
plt.figure(figsize=(10, 6))
sns.violinplot(data=product_df1, x='rating', inner='stick', palette='Set3')
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Density')
plt.show()

# Step 2: Average Price of Products (Using a Box Plot)
plt.figure(figsize=(8, 4))  # Reduced the scale for the average price plot
sns.boxplot(data=product_df1, y='price', color='skyblue')
plt.title('Average Price of Products')
plt.ylabel('Price')
plt.show()

# Step 3: Prime Eligibility Distribution (Using a Pie Chart)
prime_eligible_counts = product_df1['isPrimeEligible'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(prime_eligible_counts, labels=prime_eligible_counts.index, autopct='%1.1f%%', colors=['lightcoral', 'lightgreen'])
plt.title('Prime Eligibility Distribution')
plt.legend(title='Prime Eligibility (1-Yes, 0-No)', loc='upper right')  # Added legend
plt.show()

# Step 4: Top Categories of Products (Using a Horizontal Bar Chart)
top_categories = product_df1['Category'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_categories.plot(kind='barh', color='mediumseagreen')
plt.title('Top Categories of Products')
plt.xlabel('Frequency')
plt.ylabel('Category')
plt.show()

```

```python
# Step 2: Average Price of Products (Using a Box Plot with Custom Y-Axis Limits)
plt.figure(figsize=(8, 4))
sns.boxplot(data=product_df1, y='price', color='skyblue')
plt.title('Average Price of Products')
plt.ylabel('Price')
plt.ylim(0, 100)  # Adjust the y-axis limits as needed to clearly show the average price
plt.show()
```

```python
"""
1) Product ratings mostly fall between 4 and 5 with the largest number of products having a rating of around 4.5

2) The average price of products almost always manages to come under a 100 dollars with a few exceptions. 
When zoomed in, it was observed that most of the products under $100 fall in the range of $10-$30. 
One major exception in this trend was a product of $800 which might be a luxury beauty product. 

3) 95.5% of the products in the dataset were prime eligible. 4.5% of them were not, 
these products can be of local markets abroad since they are usually not aware of this. 

4) Surpprisingly, eye creams and clay masks are the highest number of products sold on Amazon, while body scrubs 
and cleansing balms are the lowest number of products sold there. 
"""

```

```python
# Double Variable EDA:

# Step 6: Correlation between Ratings and Prices (Using a Hexbin Plot)
plt.figure(figsize=(8, 6))
sns.jointplot(data=product_df1, x='price', y='rating', kind='hex', cmap='YlGnBu')
plt.xlabel('Price')
plt.ylabel('Rating')
plt.subplots_adjust(top=0.9)
plt.suptitle('Correlation between Ratings and Prices')
plt.show()

# Step 7: Average Ratings for Prime Eligibility (Using a Grouped Bar Chart)
avg_ratings_by_prime = product_df1.groupby('isPrimeEligible')['rating'].mean()
plt.figure(figsize=(8, 6))
avg_ratings_by_prime.plot(kind='bar', color=['lightcoral', 'lightblue'])
plt.title('Average Ratings by Prime Eligibility')
plt.xlabel('Prime Eligibility')
plt.ylabel('Average Rating')
plt.xticks(rotation=0)
plt.show()

# Step 8: Patterns between Categories and Prices (Using a Swarm Plot)
plt.figure(figsize=(12, 6))
sns.swarmplot(data=product_df1, x='price', y='Category', palette='Set2')
plt.title('Patterns between Categories and Prices')
plt.xlabel('Price')
plt.ylabel('Category')
plt.show()

# Step 9: Relationship between Total Reviews and Ratings (Using a Bubble Plot)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=product_df1, x='totalReviews', y='rating', hue='isPrimeEligible', size='price', sizes=(20, 200))
plt.title('Relationship between Total Reviews and Ratings')
plt.xlabel('Total Reviews')
plt.ylabel('Rating')
plt.legend(title='Prime Eligibility', loc='upper right')
plt.show()
```
